```python
portfolio = pd.read_csv("it_dashboard_it_portfolio.csv",
                        usecols=['Unique Investment Identifier', 'Budget Year', 'Investment Title', 'Investment Description', 'Cloud Computing Alternatives Evaluation', 'Derived Status', 'Type of Investment', 'Part Of IT Portfolio', 'FEA BRM Services - Primary service area', 'Updated Date',
                                 'O&M BY Agency Funding ($ M)',
                                 'O&M BY Budget Authority Agency Funding ($ M)',
                                 'O&M BY Contributions ($ M)',
                                 'O&M CY Agency Funding ($ M)',
                                 'O&M CY Contributions ($ M)',
                                 'O&M PY Agency Funding ($ M)',
                                 'O&M PY Contributions ($ M)',
                                 'DME BY Agency Funding ($ M)',
                                 'DME BY Budget Authority Agency Funding ($ M)',
                                 'DME BY Contributions ($ M)',
                                 'DME CY Agency Funding ($ M)',
                                 'DME CY Contributions ($ M)',
                                 'DME PY Agency Funding ($ M)',
                                 'DME PY Contributions ($ M)',
                                 'Total IT Spending FY2019 (PY) ($ M)',
                                 'Total IT Spending FY2020 (CY) ($ M)',
                                 'Total IT Spending FY2021 (BY) ($ M)', ])
```

```python
portfolio = portfolio.rename(columns={"Unique Investment Identifier": "Unique_Investment_Identifier",
                                      "Budget Year": "Budget_Year",
                                      "Investment Title": "Investment_Name",
                                      "Investment Description": "Description",
                                      "Cloud Computing Alternatives Evaluation": "Cloud_Computing_Alternatives_Evaluation",
                                      "Derived Status": "Investment_Status_Name",
                                      "Type of Investment": "Investment_Type",
                                      "Part Of IT Portfolio": "IT_Portfolio",
                                      "FEA BRM Services - Primary service area": "primary_service_area",
                                      "Updated Date": "Updated_Date"}).set_index('Unique_Investment_Identifier')
```

```python
portfolio.head()
```

```python
portfolio.dtypes
```

# Join Tables

```python
it_dashboard = portfolio.join(business_case, on='Unique_Investment_Identifier', how='left').reset_index()
it_dashboard.index.rename("Investment_Id", inplace=True)
```
